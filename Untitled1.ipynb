```python
# Function 1: Market Order
def place_market_order(symbol, side, quantity):
    try:
        order = client.futures_create_order(symbol=symbol, side=side, type=FUTURE_ORDER_TYPE_MARKET, quantity=quantity)
        print(f"Market Order Success: {order['orderId']}")
        logging.info(f"Market Order: {order}")
    except Exception as e:
        print(f"Error: {e}")

# Function 2: Limit Order
def place_limit_order(symbol, side, quantity, price):
    try:
        order = client.futures_create_order(symbol=symbol, side=side, type=FUTURE_ORDER_TYPE_LIMIT, timeInForce=TIME_IN_FORCE_GTC, quantity=quantity, price=price)
        print(f"Limit Order Success: {order['orderId']}")
        logging.info(f"Limit Order: {order}")
    except Exception as e:
        print(f"Error: {e}")

# --- Rendu order-aiyum ippo execute panrom ---
place_market_order('BTCUSDT', 'BUY', 0.001)
place_limit_order('BTCUSDT', 'BUY', 0.001, 60000)
```
